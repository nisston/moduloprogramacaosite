```python
# Definindo variáveis e solicitando entrada do usuário
nome = input("Digite seu nome: ")

# Exibindo os dados do usuário
print("\nInformações do usuário:")
print("Nome:", nome)

# Operações básicas
numero1 = int(input("\nDigite o primeiro número: "))
numero2 = int(input("Digite o segundo número: "))

soma = numero1 + numero2
print(f"\n{numero1} + {numero2} = {soma}")
```

```python
# Definindo variáveis e solicitando entrada do usuário
nome = input("Digite seu nome: ")
idade = int(input("Digite sua idade: "))
altura = float(input("Digite sua altura em metros (ex: 1.75): "))

# Exibindo os dados do usuário
print("\nInformações do usuário:")
print("Nome:", nome)
print("Idade:", idade)
print("Altura:", altura)

# Operações básicas
numero1 = int(input("\nDigite o primeiro número: "))
numero2 = int(input("Digite o segundo número: "))

# Soma
soma = numero1 + numero2
print(f"\n{numero1} + {numero2} = {soma}")

# Subtração
subtracao = numero1 - numero2
print(f"{numero1} - {numero2} = {subtracao}")

# Multiplicação
multiplicacao = numero1 * numero2
print(f"{numero1} * {numero2} = {multiplicacao}")

# Divisão
if numero2 != 0:
    divisao = numero1 / numero2
    print(f"{numero1} / {numero2} = {divisao:.2f}")
else:
    print("Divisão por zero não é permitida.")

# Exponenciação
exponenciacao = numero1 ** numero2
print(f"{numero1} ^ {numero2} = {exponenciacao}")

# Módulo
modulo = numero1 % numero2
print(f"{numero1} % {numero2} = {modulo}")
```

# Estruturas condicionais

```python
# Sintaxe - estrutura condicional simples
if condicao:
    # bloco de código
```

```python
# Exemplo prático
idade=18
if idade >= 18:
    print("Maior de idade.")
```

```python
# Sintaxe - estrutura condicional composta
if condição:
    # bloco de código se condição for verdadeira
else:
    # bloco de código se condição for falsa
```

```python
# Exemplo prático
nota = 60
if nota >= 60:
    print("Aprovado")
else:
    print("Reprovado")
```

```python
# Sintaxe - Estrutura condicional if-elif-else
if condição1:
    # bloco de código se condição1 for verdadeira
elif condição2:
    # bloco de código se condição2 for verdadeira
else:
    # bloco de código se todas as condições anteriores forem falsas
```

```python
# Exemplo prático
nota = 85
if nota >= 90:
    print("Excelente")
elif nota >= 70:
    print("Bom")
else:
    print("Precisa melhorar")
```

# Estruturas condicionais aninhadas

```python
# Estrutura condicional aninhada
# Exemplo prático
idade = 25

if idade < 18:
    print("Menor de idade")
else:
    if idade < 65:
        print("Adulto")
    else:
        print("Idoso")
```

```python
# Estrutura condicional aninhada - if-elif-else
# Exemplo prático
idade = 25
ocupacao = "estudante"

if idade < 18:
    print("Menor de idade")
    if ocupacao == "estudante":
        print("Estudante menor de idade")
elif idade < 65:
    print("Adulto")
    if ocupacao == "trabalhador":
        print("Adulto trabalhador")
    else:
        print("Adulto não trabalhador")
else:
    print("Idoso")
```

```python
valor = int(input('Informe o valor da temperatura..: '))

if valor<70:
  print('Baixo')
elif (valor>=70 and valor<85):
  print('Médio')
else:
  print('Alto')
```
