# Estrutura de repetição "for"

```python
# Estrutura for simples
for i in range(5):
    print("Contador é:", i)
```

```python
# Estrutura for com else:
for i in range(5):
    print(i)
else:
    print("Laço completo!")
```

```python
for i in range(5):
  print(i, end=' ')
```

```python
for i in range(5,15):
  print(i, end=' ')
```

```python
for i in range(5,10,2):
  print(i, end=' ')
```

# O “for” e o “while” aninhados

```python
# Lista de números
numeros = [3, 2, 1]

# Laço for interno
for i in numeros:
    indice = 0
    # Laço while externo
    while indice < len(numeros):
        num = numeros[indice]
        indice += 1
        print('Contando a partir de ',i, ' -> ',num)
```

# Comandos "break" e "continue"

```python
# Sai do laço quando i for 5
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# Pula a iteração atual se i for par
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```
