# Estrutura de repetição While

```python
# Exemplo prático
contador = 0
while contador < 5:
    print("Contador é:", contador)
    contador += 1
```

```python
# Estrutura While com if - else
contador = 0
while contador < 5:
    if contador % 2 == 0:
        print(contador, "é par")
    else:
        print(contador, "é ímpar")
    contador += 1
```

```python
# Estrutura While com else
contador = 0
while contador < 5:
    print("Contador é:", contador)
    contador += 1
else:
    print("Laço while terminou.")
```

```python
# Trabalhando a estrutura While de maneira aninhada
i = 1
while i <= 3:
    j = 1
    while j <= 3:
        print(f"i = {i}, j = {j}")
        j += 1
    i += 1
```

```python
# Definindo dois conjuntos de números
conjunto1 = [1, 2, 3]
conjunto2 = ['a', 'b', 'c']

# Trabalhando a estrutura for de maneira aninhada
for num in conjunto1:
    for letra in conjunto2:
        print(num, letra)
```

```python
# Lista de números
numeros = [3, 2, 1]

# Laço while externo
indice = 0
while indice < len(numeros):
    num = numeros[indice]

    # Laço for interno
    for i in range(num, 0, -1):
        print(f"Contando a partir de {num}: {i}")

    indice += 1
```

# Comandos break e continue

```python
# Sai do laço quando contador for 5
contador = 0
while contador < 10:
    if contador == 5:
        break
    print(contador)
    contador += 1
```

```python
# Pula a iteração atual se contador for par
contador = 0
while contador < 10:
    contador += 1
    if contador % 2 == 0:
        continue
    print(contador)
```

```python
# Exemplo prático
c = 1
soma=0
while c <= 3:
    valor = int(input('Informe um valor..: '))
    c += 1
    soma+=valor
print('A media foi de...: ',(soma/3))
```

```python
# Exemplo prático
c = 1
menor=1000
while c <= 3:
    valor = int(input('Informe um valor..: '))
    if valor<menor:
      menor=valor
    c += 1

print('O menor valor informado foi de...: ',menor)
```

```python
# Exemplo prático
c=1
sp=0
si=0
while c <= 1000:
    if c%2==0:
      sp+=c
    else:
      si+=c
    c += 1

print('A soma dos pares foi de...: ',sp)
print('A soma dos pares foi de...: ',si)
```
