# Introdução às Strings


```python
# Introdução às Strings
# Strings em Python são sequências de caracteres usadas para armazenar e manipular texto.
# Elas são delimitadas por aspas simples ('') ou duplas ("").

# Exemplo 1
string1 = "Olá, Mundo!"
print(string1)
```

```python
# Exemplo 2
string2 = 'Python é incrível!'
print(string2)
```

# Operações Básicas com Strings

```python
# --------------------------------------------

# Operações Básicas com Strings
# Podemos concatenar, repetir e acessar caracteres individuais usando índices.

# Exemplo 1: Concatenando Strings
string1 = "Olá"
string2 = "Mundo"
resultado = string1 + ", " + string2 + "!"
print(resultado)
```

```python
# Exemplo 2: Repetindo Strings
string3 = "Python! " * 3
print(string3)
```

# Manipulação de Strings

```python
# --------------------------------------------

# Manipulação de Strings
# É possível usar fatiamento e verificar a presença de caracteres ou palavras.

# Exemplo 1: Fatiamento
texto = "Manipulação"
print(texto[0:5])  # Saída: Manip
print(texto[-3:])  # Saída: ção
```

```python
# Exemplo 2: Verificação de Substring
frase = "Python é ótimo"
print("ótimo" in frase)  # Saída: True
print("difícil" not in frase)  # Saída: True
```

# Formatando Strings

```python
# --------------------------------------------

# Formatando Strings
# O método `format` e f-strings são usados para inserir valores dentro de strings.

# Exemplo 1: Usando `format`
nome = "Alice"
idade = 25
mensagem = "Meu nome é {} e tenho {} anos.".format(nome, idade)
print(mensagem)
```

```python
# Exemplo 2: Usando f-strings
mensagem_fstring = f"Meu nome é {nome} e tenho {idade} anos."
print(mensagem_fstring)
```

# Strings Multilinhas

```python
# --------------------------------------------

# Strings Multilinhas
# Strings multilinhas são criadas usando três aspas simples ou duplas.

# Exemplo 1: Multilinha com três aspas duplas
texto = """Este é um exemplo
de string
multilinha."""
print(texto)
```

```python
# Exemplo 2: Multilinha com três aspas simples
outro_texto = '''Outra string
multilinha em Python.'''
print(outro_texto)
```

# Strings como Iteráveis

```python
# --------------------------------------------

# Strings como Iteráveis
# Podemos iterar sobre os caracteres de uma string usando loops.

# Exemplo 1: Iteração com `for`
palavra = "Python"
for letra in palavra:
    print(letra)
```

```python
# Exemplo 2: Usando compreensão de listas
letras = [letra for letra in palavra]
print(letras)
```

# Métodos de Divisão e Junção

```python
# --------------------------------------------

# Métodos de Divisão e Junção
# Strings podem ser divididas em partes ou unidas a partir de listas.

# Exemplo 1: Dividindo Strings
frase = "Dividindo esta frase"
partes = frase.split(" ")
print(partes)
```

```python
# Exemplo 2: Juntando Strings
nova_frase = "-".join(partes)
print(nova_frase)
```

# Escape Characters e Raw Strings

```python
# --------------------------------------------

# Escape Characters e Raw Strings
# Escape characters permitem incluir caracteres especiais.
# Raw strings desativam escapes.

# Exemplo 1: Escape Characters
frase_com_aspas = "Ele disse: \"Python é incrível!\""
print(frase_com_aspas)

# Exemplo 2: Raw Strings
raw_string = r"C:\novo_diretorio\arquivo.txt"
print(raw_string)
```
